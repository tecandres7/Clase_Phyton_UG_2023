```python
import mysql.connector
import random
import string
from datetime import date

# Configuración de la conexión a la base de datos
cnx = mysql.connector.connect(user='user', password='password',
                              host='localhost', database='libro_curso')
cursor = cnx.cursor()

# Función para generar códigos de curso aleatorios
def generar_codigo_curso():
    letras = ''.join(random.choices(string.ascii_uppercase + string.digits, k=2))
    semestre = random.randint(1, 10)
    codigo = f"{letras}-{semestre:02}"
    return codigo

# Función para generar códigos de ISBN aleatorios
def generar_ISBN():
    digitos = ''.join(random.choices(string.digits, k=11))
    codigo = f"2{digitos}"
    return codigo

# Generación de 47 cursos
for i in range(47):
    codigo = generar_codigo_curso()
    nombre = ''.join(random.choices(string.ascii_uppercase + string.digits, k=25))
    descripcion = ''.join(random.choices(string.ascii_uppercase + string.digits, k=100))
    query = f"INSERT INTO curso (codigo, nombre, descripcion) VALUES ('{codigo}', '{nombre}', '{descripcion}')"
    cursor.execute(query)
    cnx.commit()

# Selección de todos los cursos
query = "SELECT id FROM curso"
cursor.execute(query)
cursos = cursor.fetchall()

# Generación de 150 libros
for i in range(150):
    nombre = ''.join(random.choices(string.ascii_uppercase + string.digits, k=50))
    autor = ''.join(random.choices(string.ascii_uppercase + string.digits, k=45))
    edicion = random.randint(4, 5)
    editorial = ''.join(random.choices(string.ascii_uppercase + string.digits, k=40))
    ano = random.randint(2000, date.today().year)
    ISBN = generar_ISBN()
    curso_id = random.choice(cursos)[0]
    query = f"INSERT INTO libro (ISBN, nombre, autor, edicion, editorial, año, curso_id) VALUES ('{ISBN}', '{nombre}', '{autor}', {edicion}, '{editorial}', {ano}, {curso_id})"
    cursor.execute(query)
    cnx.commit()

# Cierre de la conexión a la base de datos
cursor.close()
cnx.close()
```

```python
import random
import string

# Lista de editoriales
editoriales = ['Editorial A', 'Editorial B', 'Editorial C', 'Editorial D', 'Editorial E']

# Función para generar códigos ISBN
def generate_ISBN():
    prefix = ''.join(random.choice(string.digits[2:]) for _ in range(3))
    identifier = ''.join(random.choice(string.digits) for _ in range(9))
    check_digit = str((10 - sum(int(digit) * (i % 2 * 2 + 1) for i, digit in enumerate(prefix + identifier))) % 10)
    return prefix + identifier + check_digit

# Generar 150 libros
for i in range(1, 151):
    ISBN = generate_ISBN()
    nombre = f'Libro {i}'
    autor = ''.join(random.choice(string.ascii_uppercase + string.ascii_lowercase + ' ') for _ in range(random.randint(5, 15)))
    edicion = random.randint(4, 5)
    editorial = random.choice(editoriales)
    year = random.randint(2000, 2023)
    curso_id = random.randint(1, 47)
    print(f"INSERT INTO libro(ISBN, nombre, autor, edicion, editorial, anio, curso_id) VALUES ('{ISBN}', '{nombre}', '{autor}', {edicion}, '{editorial}', {year}, {curso_id});")
    #print(f"INSERT INTO libro(ISBN, nombre, autor, edicion, editorial, anio) VALUES ('{ISBN}', '{nombre}', '{autor}', {edicion}, '{editorial}', {year});")
```
