##### 4. Dados los arrays a = [1, 4, 2, 7] y b = [1, 3, 2, 9], obtenga la media aritmética de la diferencia (a-b).

```python

a = [1, 4, 2, 7]
b = [1, 3, 2, 9]

diferencia = np.subtract(a, b)
media = np.mean(diferencia)

print("La media aritmética de la diferencia es:", media)
```

```python
# Seleccionar el municipio situado más al norte
municipio_norte = df.loc[df["Y"].idxmax()]
print(municipio_norte)
print("")
# Obtener el nombre del municipio, la provincia y la población
nombre_municipio_norte = municipio_norte["Municipio"]
provincia_municipio_norte = municipio_norte["Provincia"]
poblacion_municipio_norte = municipio_norte["Poblacion"]

print("El municipio situado más al norte es", nombre_municipio_norte, "que pertenece a la provincia de", provincia_municipio_norte, "y tiene una población de", poblacion_municipio_norte, "habitantes.")
```

##### 5) Obtener estadísticas de la población de los municipios de la provincia de Granada


```python
poblacion_granada = df.loc[df["Provincia"] == "Granada"]["Poblacion"]
media_poblacion_granada = poblacion_granada.mean()
mediana_poblacion_granada = poblacion_granada.median()
desviacion_poblacion_granada = poblacion_granada.std()
max_poblacion_granada = poblacion_granada.max()
min_poblacion_granada = poblacion_granada.min()
print("Estadísticas de población para los municipios de Granada:")
print("Media:", media_poblacion_granada)
print("Mediana:", mediana_poblacion_granada)
print("Desviación estándar:", desviacion_poblacion_granada)
print("Valor máximo:", max_poblacion_granada)
print("Valor mínimo:", min_poblacion_granada)
```

##### 6) Realizar histograma de la población de los Municipios para cada una de las provincias


```python
provincias = df["Provincia"].unique()
for provincia in provincias:
    poblacion = df.loc[df["Provincia"] == provincia]["Poblacion"]
    plt.hist(poblacion, bins=50)
    plt.title("Histograma de población para " + provincia)
    plt.xlabel("Población")
    plt.ylabel("Frecuencia")
    plt.show()
```
