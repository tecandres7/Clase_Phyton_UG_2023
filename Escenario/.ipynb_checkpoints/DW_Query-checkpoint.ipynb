####  Hacemos el Select de Clientes y Generos de todos los elementos para hacer el dF

```python
sql_query= 'SELECT * FROM clientes'
df_clientes = pd.read_sql(sql_query, postgres_driver)
df_clientes

sql_query= 'SELECT * FROM generos'
df_generos = pd.read_sql(sql_query, postgres_driver)
df_generos

df_clientes
```

####  Hacer un merge o Joint entre la tabla clientes_generos

```python
df_clientes_generos = df_clientes.merge(df_generos,left_on='id_genero',
                                        right_on='id_genero',how='inner')
df_clientes_generos
```

```python
sql_query = 'SELECT * FROM direcciones'
df_direcciones = pd.read_sql(sql_query, postgres_driver)
dimClientes = df_clientes_generos.merge(df_direcciones,
                                        left_on='id_direccion', right_on='id_direccion', how='inner')
dimClientes.drop(['id_direccion','id_genero'], axis=1, inplace = True)
dimClientes
```

```python
sql_query = 'SELECT * FROM sectores'
df_sectores = pd.read_sql(sql_query, postgres_driver)
dimClientes = dimClientes.merge(df_sectores,
                        left_on='id_sector', right_on='id_sector', how='inner')
dimClientes.drop(['id_sector','latitud','longitud'], axis=1, inplace=True)
dimClientes
```
