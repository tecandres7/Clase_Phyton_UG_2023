```python


continuas = []
discretas = []
categoricas = []
for col in df3.columns.values:
    if((df[col].dtype=='int64') or (df[col].dtype=='float64')):
            cantidad = len(df3[col].value_counts())
            if(cantidad>=30):
                continuas.append(col)
            else:
                discretas.append(col)
    else:
        categoricas.append(col)
return continuas, discretas, categoricas
```

```python
dataset = pd.read_csv('hotel_bookings.csv')
dataset.shape
```

```python
dataset.head(100)
```

```python
##  Que sean de francia y que no sean niños y que no sean babies
dataset[(dataset['country']=='FRA')&
        (dataset['children']==0)&
        (dataset['babies']==0)]
```

```python
###Hacer grupos que se le pueden aplicar algún tipo de agregación

group_country = dataset.groupby('country')
group_country.sum()
```

```python
#Cantidad 
dataset['hotel'].unique()
dataset['hotel'].value_counts()
```

```python
## Agregación por grupo 

dataset.groupby('hotel').agg(
    media = ('adults', np.mean),
    suma = ('adults', np.sum),
    media_children = ('children', np.mean),
    suma_children = ('children', np.sum)
)
```

```python
##  Gráfica de Densidad

dataset['adults'].plot(kind='density')
dataset['adults'].describe()
```
